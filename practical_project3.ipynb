```python
import nltk
from  nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
bbc = pd.read_csv('bbc_news.csv')
bbc.head()
```

```python
bbc.info()
```

```python
title = pd.DataFrame(bbc['title'])
title.head()
```

```python
#clean data
#lowercase
title['lower'] = title['title'].str.lower()
title.head()
```

```python
nltk.download('stopwords')
```

```python
#stopwords
en_stopwords = stopwords.words('english')
title['no_stopwords'] = title['lower'].apply(lambda x:" " .join([word for word in x.split() if word not in(en_stopwords)]))
title.head()
```

```python
title["no_stopwords_no_punct"] = title.apply(
    lambda x: re.sub(r"[^\w\s]","" ,x['no_stopwords']),
    axis=1
)
title.head()
```

```python
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt_tab')
```

```python
#tokenize
title["tokenize"] = title.apply(lambda x: word_tokenize(x['no_stopwords_no_punct']), axis = 1)
title.head()
```

```python
#lemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
title["lemmatised"] = title["tokenize"].apply(lambda tokens: [ lemmatizer.lemmatize(token) for token in tokens])
title.head()
```

```python
token_new_list = sum(title['tokenize'] , [])
token_clean_list = sum(title['lemmatised'] , [])
```

```python
token_new_list
```

```python
token_clean_list
```

```python
#pos tagging
nlp = spacy.load('en_core_web_sm')
```

```python
spacy_doc = nlp(" ".join(token_new_list))
```

```python
pos_df = pd.DataFrame(columns=['tokens' ,'pos_tag'])
for token in spacy_doc:
    pos_df = pd.concat([pos_df ,pd.DataFrame.from_records([{'token' : token.text , "pos_tag" :token.pos_}])],
                        ignore_index=True)
    
```

```python
pos_df_counts = pos_df.groupby(['token' ,'pos_tag']) .size().reset_index(name ='counts').sort_values(
    by='counts' ,ascending = False)
pos_df_counts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts['pos_tag']== 'noun'][0:10]
nouns
```

```python
#ner
ner_df = pd.DataFrame(columns=['token' ,'ner_tag'])
for token in spacy_doc.ents:
    if pd.isna(token.label_) is False:
        ner_df = pd.concat([ner_df , pd.DataFrame.from_records([{'token' : token.text , "pos_tag" :token.pos_} ,
                                                                ignore_index = True])])
```
